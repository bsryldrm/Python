```python
for x in range(10):
  print(x**2)
```

```python
numbers = [x**2 for x in range(10)]
print(numbers)
```

```python
numbers = [x**2 for x in range(10) if x%3 == 0]
print(numbers)
```

```python
myString = 'hello'
myList = []
```

```python
for letter in myString:
  myList.append(letter)
print(myList)
```

```python
myList = [letter for letter in myString]
print(myList)
```

```python
years = [1993, 1996, 1998, 2001]
ages = [2019-year for year in years]
print(ages)
```

```python
results = [x if x%2 ==0 else 'TEK' for x in range(1,10)]
print(results)
```

```python
results = []
for x in range(3):
  for y in range(3):
    results.append((x,y))
```
