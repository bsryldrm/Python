gönderilen sayının karesini al



```python
def square(num): return num **2
```

```python
result = square(2)
```

elimizde bir liste varsa ve listenin her bir elemanının karesini alıp ekrana yazmak istersek

```python
# map()
numbers = [ 1,3,5,9]

result = list(map(square, numbers))
print(result)
```

```python
for item in map(square, numbers):
  print(item)
```

sadece bir kereliğine ihtiyaç duyacağımız bir fonksiyon inşa etmek istersek

```python
# lambda
numbers = [1,3,5,9]

result = list(map(lambda num: num ** 2, numbers))
print(result)
```

```python
# lambda
square = lambda num: num ** 2
numbers = [1,3,5,9]

result = list(map(square, numbers))
print(result)
```

```python
result = square(3)
print(result)
```

döndürülen result işlemi üzerinde bir filtreleme yapmak istersek

```python
#sayıların sadece 3 e bölünenler dönsün
# result = [1, 9, 25, 81]

result2 = list(filter(lambda num : num % 3 ==0, result))
print(result2)
```
