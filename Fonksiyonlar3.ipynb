lokal ve global değişkenler

```python
x = 'global x'

def function():
  x = 'local x'
  print(x)

function()
```

```python
print(x)
```

fonksiyon içinde tanımlı bir x değişkeni olduğu için fonksiyonu çağırdığımızda onu yazar. ama fonksiyondan hariç print ile x değişkeni yazdırmak istediğimizde global olan x değişkendeki ifadeyi yazar

```python
x = 'global x'

def function():
  #x = 'local x'
  print(x)

function()
```

fonksiyon içinde tanımlı bir değişken yoksa fonksiyonu çağırdığımızda dahi global olarak tanımlı x değişkendki değeri yazar.

```python
name = 'global string'

def greeting():
  name = 'Çınar'
  def hello():
    print('Hello ' + name)

  hello()

greeting()
```

```python
x = 50
def test():
  global x
  print(f'x : {x}')

  x = 100
  print(f"changed x to {x}")

test()
```
