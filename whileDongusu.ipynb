1- Sayilar listesini while ile ekrana yazdırın

```python
i = 0
while(i< len(sayilar)):
  print(sayilar[i])
  i+= 1
```

2- Başlangıç ve bitiş değerlerini kullanıcıdan alıp aradaki tüm tek sayıları ekrana yazdırın

```python
baslangic = int(input('baslangıç:'))
bitis = int(input('bitis:'))

i = baslangic
while i < bitis:
  if i % 2 == 1:
    print(i)
  i += 1
```

4- kullanıcıdan alacağınız 5 sayıyı ekranda sıralı bir şekilde yazdırın.

```python
numbers = []
i = 0

while i < 5:
  numbers.append(int(input('sayi giriniz : ')))
  i += 1
numbers.sort()
```

5- kullanıcıdan alacağınız sınırsız ürün bilgisini urunler listesi içersine saklayınız

 ** ürün sayısını kullanıcıya sorun

 ** dictionary listesini yapısı (name, price) şekliinde olsun.


```python
urunler = []

adet = int(input('kaç ürün eklemek istiyorsunuz: '))
i = 0

while(i < adet):
  name = input('ürün ismi: ')
  price = input('ürün fiyatı: ')
  urunler.append({
      'name' : name ,
      'price' : price
  })

  i += 1
```
